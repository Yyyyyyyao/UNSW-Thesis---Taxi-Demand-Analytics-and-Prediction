```python
import pandas as pd
import numpy as np
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
new_df = pd.read_csv('yellow_tripdata_2017-02.csv', parse_dates=['tpep_pickup_datetime'], date_parser=dateparse)

df = new_df[['tpep_pickup_datetime', 'PULocationID']]

df['Hours'] = new_df['tpep_pickup_datetime'].dt.hour
df['Date'] = new_df['tpep_pickup_datetime'].dt.date

test_df = pd.DataFrame({'count' : df.groupby( ['PULocationID', 'Date', 'Hours'] ).size()}).reset_index()
test_df
```

```python
test_df.to_csv('xg_2017_02_step1.csv', index=False)
```
