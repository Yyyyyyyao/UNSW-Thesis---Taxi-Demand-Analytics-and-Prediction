```python
import pandas as pd
new_df = pd.read_csv('xg_2017_02_dataset.csv')
```

```python
new_df['count'].max()
```

```python
new_df
```

```python
new_df['dayofweek'] = new_df['dayofweek'] / 6
```

```python
new_df['Hours'] = new_df['Hours'] / 23
new_df['time1'] = new_df['time1'] / new_df['time1'].max()
new_df['time2'] = new_df['time2'] / new_df['time2'].max()
new_df['time3'] = new_df['time3'] / new_df['time3'].max()
```

```python
new_df['count'] = new_df['count'] / new_df['count'].max()
```

```python
new_df['PULocationID'] = new_df['PULocationID'] / new_df['PULocationID'].max()
```

```python
new_df.to_csv('normalized_data_2017_02.csv', index=False)
```
