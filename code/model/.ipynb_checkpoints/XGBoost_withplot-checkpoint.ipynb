```python
#new_df = pd.read_csv('xg_2017_01_final.csv')

new_df = pd.read_csv('../normalized_dataset/xg_2017_0102Combine_Normalization.csv')
#new_df = pd.read_csv('../normalized_data_2017_01.csv')
# X_train, y_train = new_df.iloc[:, :-1], new_df.iloc[:, -1]

X, y = new_df.iloc[:, :-1], new_df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)


# data_dmatrix = xgb.DMatrix(data=X,label=y) 
# default learning rate = 0.1186
# default objective = 'reg:linear'

xg_reg = xgb.XGBRegressor(objective='reg:logistic', colsample_bytree = 0.9, 
                          subsample= 0.96, min_child_weight = 6.19, learning_rate = 0.3, 
                          max_depth=14, gamma=1, n_estimators = 463)

# Fit data into the XGBoost model
xg_reg.fit(X_train,y_train)

# Make predictions
preds = xg_reg.predict(X_test)

y_test = y_test*2149
preds = preds*2149

rmse = np.sqrt(mean_squared_error(y_test[:], preds[:]))
print("Test RMSE: %f" % (rmse))

testacc = statistics.mean(abs(y_test[:] - preds[:]))
print('Test MAE: %.2f' % (testacc))

# mean_percentage_error = statistics.mean(abs(y_test[:] - preds[:])/y_test[:])

mean_percentage_error = np.sum(abs(y_test[:] - preds[:]))/np.sum(y_test[:])

print('Test PE: %.2f' % (mean_percentage_error))

plt.plot(range(100),preds[0:100], c='r')
plt.plot(range(100),y_test[0:100], c='g')
plt.show()

```
